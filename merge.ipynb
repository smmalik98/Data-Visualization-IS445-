```python

#reading datasets
df1 = pd.read_csv('clean_crashes.csv') #crashes
df2 = pd.read_csv('clean_vehicle.csv') #vehicles
df3 = pd.read_csv('clean_people.csv') #people
df4 = pd.read_csv('clean_red.csv') #red_light violation
df5 = pd.read_csv('clean_speed.csv') #speed camera violation
```

# cleaning data

```python
df1=df1.loc[(df1['LATITUDE']!=0) | (df1["LATITUDE"]!=0)]
```

```python
df3=df3.loc[df3['SEX']!='X']
```

```python
df2=df2.loc[df2['MODEL']!='UNKNOWN']
```

```python
#using particular columns only

df3=df3[['CRASH_RECORD_ID', 'VEHICLE_ID', 'DATE', 'PERSON_TYPE', 'SEX', 'AGE',
       'SAFETY_EQUIPMENT', 'AIRBAG_DEPLOYED', 'INJURY_CLASSIFICATION','year']]

df2=df2[['VEHICLE_ID', 'UNIT_TYPE', 'MAKE', 'MODEL','VEHICLE_TYPE', 'VEHICLE_USE','year']]

df1=df1[['CRASH_RECORD_ID', 'STREET_NAME', 'LATITUDE', 'LONGITUDE','POSTED_SPEED_LIMIT', 'WEATHER_CONDITION', 'LIGHTING_CONDITION','ROADWAY_SURFACE_COND', 'DAMAGE', 'year', 'month', 'day_of_month','hour', 'minute', 'second', 'day']]


```

```python
df1 = df1.dropna(axis = 0, how ='any')
df2 = df2.dropna(axis = 0, how ='any')
df3 = df3.dropna(axis = 0, how ='any')
```

```python

t=[]
for i in list(range(2015,2023)):
    df2_x=df2.loc[df2['year']==(i)]
    df3_x=df3.loc[df3['year']==(i)]
    df1_x=df1.loc[df1['year']==(i)]
    
    merge_d23=pd.merge(df2_x,df3_x, how='inner',on='VEHICLE_ID').rename(columns={'CRASH_RECORD_ID_x': 'CRASH_RECORD_ID'}) #merge people and vehicle
    
    merge_d31=pd.merge(merge_d23,df1_x, how='inner',on='CRASH_RECORD_ID')
    print(merge_d31.shape,i)
    t.append(merge_d31)

    merge_d31.to_csv('year_data/'+str(i)+'.csv', index=False)
```

```python
merge_dataset=pd.concat(t)
```

```python
#drop columns
merge_dataset = merge_dataset.drop(columns=['year_x', 'year_y'])
```
